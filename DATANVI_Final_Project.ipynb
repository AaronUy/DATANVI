```python
# Get data
rate = pd.read_csv('/content/drive/MyDrive/DATANVI/SEA Quarterly GDP Growth Rate.csv')
case = pd.read_csv('/content/drive/MyDrive/DATANVI/SEA Quarterly Confirmed COVID-19 Cases.csv')
```

 **For this part, we wanted to look at the GDP Growth Rate and number of active COVID-19 cases of the SEA countries.
We listed the SEA countries in countryNames, then we listed the 4th quarter of 2019 (when COVID-19 began) up to the 1st quarter of 2021 (the latest data we have) in the quarters variable**

```python
rate = rate.drop(columns="Country")
case = case.drop(columns="Country")

countryNames = ["Brunei", "Burma (Myanmar)", "Cambodia", "Indonesia", "Laos", 
                "Malaysia", "Philippines", "Singapore", "Thailand", "Timor Leste", "Vietnam"]
quarters = ["2019-4", "2020-1", "2020-2", "2020-3", "2020-4", "2021-1"]
```

**Listed below is the growth rate of the SEA countries per quarter. The countries are listed alphabetically. 0 = Brunei ..... 10 = Vietnam**

**Below here is the number of COVID-19 cases of the SEA countries per quarter. The countries are also listed alphabetically.**

```python
case
```

```python
rate_t = rate.transpose()
rate_t
```

```python
case_t = case.transpose()
case_t
```

**In this part, we plotted the results above (GDP Growth Rate and COVID-19 Cases) of each SEA country into a line graph for better visualization. The vis can still be improved.**

```python
fig, axs = plt.subplots(11, 2, figsize=(15,15))
for i in range(11):
  rate_t[i].plot(ax=axs[i, 0])
  axs[i, 0].set_title(countryNames[i] + " GDP Growth Rate", fontsize=15)
  case_t[i].plot(ax=axs[i, 1])
  axs[i, 1].set_title(countryNames[i] + " Confirmed COVID-19 Cases", fontsize=15)

fig.tight_layout()
```

```python
case[quarters[2]]
```

**For this part, we looked into the GDP Growth Rate and COVID-19 Caeses for each quarter. We placed them into a bar chart in order to compare the SEA countries with one another. The vis can still be improved.**

```python
fig, axs = plt.subplots(6, 2, figsize=(15,15))
for i in range(6):
  rate[quarters[i]].plot(ax=axs[i, 0], kind="bar")
  axs[i, 0].set_xticklabels(countryNames)
  axs[i, 0].set_title(quarters[i] + " GDP Growth Rate", fontsize=15)
  case[quarters[i]].plot(ax=axs[i, 1], kind="bar")
  axs[i, 1].set_xticklabels(countryNames)
  axs[i, 1].set_title(quarters[i] + " Confirmed COVID-19 Cases", fontsize=15)

fig.tight_layout()
```

**For this part, we placed all of the SEA countries into one graph each to compare their GDP Growth Rate and COVID-19 Cases.**

**On the GDP graph, every SEA country's GDP went lower in between the 4th quarter of 2019 and 1st quarter of 2020 due to COVID-19. Philippines is last place in GDP Growth Rate as of the 1st quarter of 2021.**

**On the COVID-19 Cases graph, the current graph is in millions. Currently, the highest number of cases is the Red color or #3 refers to Indonesia, followed by Pink or #6 which is Philippines.**
